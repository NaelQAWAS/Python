```python
import pandas as pd
import seaborn as sns
```

```python
sns.set(context="talk", style="white")
```

```python
ebola = pd.read_csv("https://github.com/cmrivers/ebola/raw/master/country_timeseries.csv")
ebola.head()
```

```python
cases = ebola.filter(regex="Date|Cases")
cases = pd.melt(cases, "Date", var_name="Country", value_name="Cases")
print "Cases \n"
print cases.head()
cases.Country = cases.Country.str[6:]
deaths = ebola.filter(regex="Date|Deaths")
deaths = pd.melt(deaths, "Date", var_name="Country", value_name="Deaths")
print "Deaths \n"
print deaths.head()
deaths.Country = deaths.Country.str[7:]
```

```python
ebola = pd.merge(cases, deaths, on=["Date", "Country"])
ebola.Date = pd.to_datetime(ebola.Date)
```

```python
ebola.head()
```

```python
def plot(x, y, data=None, label=None, **kwargs):
    data.plot(x, y, label=label, rot=0, lw=3, grid=False, **kwargs)
```

```python
g = sns.FacetGrid(ebola, hue="Country", size=7,
                  hue_order=["Liberia", "SierraLeone", "Guinea", "Nigeria"],
                  legend_out=False)
g.map_dataframe(plot, "Date", "Cases")
ax = g.axes[0, 0]
ax.set_title("Ebola Cases in West Africa", size=14)
g.fig.tight_layout()
ax.legend(loc=(.8, .05))
#g.savefig("/Users/mwaskom/Dropbox/Temp/ebola.png")
```
